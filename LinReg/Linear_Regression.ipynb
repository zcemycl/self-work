```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = []
for n in range(10):
    a.append(random.randint(1,101))
```

```python
print(a)
```

```python
m = 1
c = 2
y = []
x = []
for n in range(10):
    y.append(m*n+c)
    x.append(n+1)
print(y)
print(x)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
p = ax.plot(a)
g = ax.plot(y)
ax.set_xlabel('index')
ax.set_ylabel('values')
ax.set_title('Plot of linear regression')
```

# Cost Function
# $ J(\theta) = \sum^{k}_{i = 1} (y^{(i)} - d^{(i)})^2
 = \displaystyle\sum_{i = 1}^{k} (mx^{(i)}+c - d^{(i)})^2$
# Slope Equation
# $ m (slope) = \frac{\bar{x}\space\cdot\space\bar{y} - \overline{xy}}{(\bar{x})^2 - \bar{x^2}} $
# $ c (intercept) = \bar{y} - m\bar{x}$

# Training set

```python
n = len(a)
y_sum = sum(a)
print(n,y_sum)
x_sum = sum(x)
x2_sum = sum(np.square(x))
xy = np.dot(x,y)
print(x_sum, x2_sum, xy)
```

```python
m_slope = (x_sum*y_sum - xy)/((x_sum)^2 - x2_sum)
print(m_slope)
c_intercept = (y_sum - m_slope*x_sum)/n
print(c_intercept)
```

```python
y_prime = []
for n in range(10):
    y_prime.append(m_slope*n+c_intercept)
fig = plt.figure()
ax = fig.add_subplot(111)
p = ax.plot(a)
g = ax.plot(y_prime)
ax.set_xlabel('index')
ax.set_ylabel('values')
ax.set_title('Plot of linear regression')
```
