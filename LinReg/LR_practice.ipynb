```python
import random
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from statistics import mean
style.use('fivethirtyeight')

# how much number created (hm), are they correlated
def create_dataset(hm, variance, step=2, correlation=False):
    val = 1 # starting value for y
    ys = [] # empty list
    for i in range(hm):
        y = val + random.randrange(-variance, variance)
        ys.append(y)
        if correlation and correlation == 'pos':
            val+= step
        elif correlation and correlation == 'neg':
            val-= step
    xs = [i for i in range(len(ys))]
    
    return np.array(xs, dtype=np.float64), np.array(ys, dtype=np.float64)
```

```python
def ideal_regression(xs,ys):
    m = (mean(xs)*mean(ys) - mean(xs*ys))/ ((mean(xs))**2 - mean(xs**2))
    b = mean(ys) - m*mean(xs)
    return m,b
```

# Positive relation

```python
xsp, ysp = create_dataset(40, 40, 2, correlation= 'pos')
mp, bp = ideal_regression(xsp, ysp)
regression_line_p = [ mp*x + bp for x in xsp] 
plt.scatter(xsp, ysp)
plt.plot(xsp,regression_line_p)
```

# Negative relation

```python
xsn, ysn = create_dataset(40, 40, 2, correlation= 'neg')
mn, bn = ideal_regression(xsn, ysn)
regression_line_n = [mn*x + bn for x in xsn]
plt.plot(xsn, regression_line_n)
plt.scatter(xsn, ysn)
```

# No trend

```python
xs, ys = create_dataset(40, 40, 2)
m, b = ideal_regression(xs, ys)
regression_line = [m*x + b for x in xs]
plt.plot(xsn, regression_line)
plt.scatter(xs, ys)
```
