# How to draw plots in python (Testing)

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')

soa = np.array([[0, 0, 3, 2], [0, 0, 1, 1], [0, 0, 9, 9]])

X, Y, U, V = zip(*soa)
print(X, Y, U, V)
plt.figure()
ax = plt.gca()
ax.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1, label= ['a','b','u','v'])

xs = [1,2]
ys = [7,9]
plt.plot(xs,ys, 'r--', label= 'projection')
ax.quiver(xs,ys, color = 'r', angles='xy', scale_units='xy', scale=1)

plt.legend()
ax.set_xlim([-1, 10])
ax.set_ylim([-1, 10])
plt.draw()
plt.show()
```

# You can see [0,0,3,2] it represents the vector tails from [0,0] and heads towards [3,2]
# The function of zip transforms those vectors into arrays of x-tail, y-tail, x-head and y-head

# Theory of Support Vector Machine (SVM)

```python
dataset = {'_':[[1,2],[2,3],[3,1]],'+':[[6,5],[7,7],[8,6]]}
new_features = [4,3]
[[plt.scatter(ii[0],ii[1], s=100, marker=i, color = 'k') for ii in dataset[i]] for i in dataset]
plt.scatter(new_features[0], new_features[1], s=100, color = 'r' , label = 'unknown class')

# decision boundary
dx = [3,6]
dy = [10,0]
plt.plot(dx,dy, 'b--', label= 'Decision Boundary')


# vectors 
# the perpendicular vector of the boundary and the vector of the unknown datapoint
plt.quiver(4.5,5.625,color = 'b', angles='xy', scale_units='xy', scale=1, label = 'boundary vector $\overrightarrow{w}$')
plt.quiver(new_features[0], new_features[1], color = 'r', angles='xy', scale_units='xy', scale=1, label = 'unknown vector $\overrightarrow{u}$')
# the projection


# legends
plt.legend(loc='upper right', bbox_to_anchor=(1.5,1))
```

# $$ \overrightarrow{u}\cdot\overrightarrow{w} + b \begin{cases}
> 0, then = \oplus \\
< 0, then = \ominus \\ 
= 0, decision\ boundary
\end{cases} , where \space b\space = bias $$
# $$ \overrightarrow{x_{\mp sv}}\cdot\overrightarrow{w} + b = \mp1  $$
# Introduce $y_i$ as classes, 
# $$ y_i = \begin{cases}
+1 \space for \space \oplus \\
-1 \space for \space \ominus  \end{cases} $$
# $For \space \oplus \space class$, i.e. $y_i$ = +1
# $$ \overrightarrow{x_i}\cdot\overrightarrow{w} + b = 1 \\
 y_i(\overrightarrow{x_i}\cdot\overrightarrow{w} + b) = 1 \\
 x_+ = 1-b$$
# $For \space \ominus \space class$, i.e. $y_i$ = -1
# $$ \overrightarrow{x_i}\cdot\overrightarrow{w} + b = -1 \\
 y_i(\overrightarrow{x_i}\cdot\overrightarrow{w} + b) = 1 \\
 x_{-} = 1+b$$
# Generally, to satisfy the hyperplane condition, $$y_i(\overrightarrow{x_i}\cdot\overrightarrow{w} + b) - 1 = 0 \rightarrow Hyperplane $$

```python
dataset_hyper = {'_':[[2,2],[0,8]],'+':[[6,2],[4,8]]}
[[plt.scatter(ii[0],ii[1], s=100, marker=i, color = 'k') for ii in dataset_hyper[i]] for i in dataset_hyper]

coordinates = []
for sign in dataset_hyper:
    for coor in range(len(dataset_hyper[sign])):
        coordinates.append(dataset_hyper[sign][coor])

print(coordinates)
coordinates = np.array(coordinates)
print(coordinates)
# hyperplane lines
A,B= zip(*coordinates)
print(A,B)
A1 = (A[0],A[1])
A2 = (A[2],A[3])
print(A1,A2)
B1 = (B[0],B[1])
B2 = (B[2],B[3])
plt.plot(A1,B1, label = 'hyperplane -')
plt.plot(A2,B2, label = 'hyperplane +')

# decision boundary
plt.plot((4,2),(2,8),label = 'decision boundary')

# width
plt.plot((1,4.3),(5,7), 'k', label = 'width')
# legend
plt.legend(loc='upper right', bbox_to_anchor=(1.5,1))
```

# Each marker represents a support vector, since the location of the marker affects the vector.

# $$ Width = (\overrightarrow{x_+} - \overrightarrow{x_-})\cdot \frac{\overrightarrow{w}}{\|\overrightarrow{w}\|} \\
= \frac{2}{\|\overrightarrow{w}\|}$$
# Want to max width and need to min $\|\overrightarrow{w}\|$ ~ min $\frac{1} {2}\|\overrightarrow{w}\|^2$ for mathematical convenience with the constraint of $y_i(x_iw + b)-1$

# Lagrange factor
# $$L(w,b) = \frac{1}{2}\|\overrightarrow{w}\|^2 - \displaystyle\sum^n_{i=1}\alpha_i(y_i(\overrightarrow{x_i}\cdot\overrightarrow{w}+b)-1) \\, where\ min\ w\ and\ max\ b$$
# So, $\frac{\partial L}{\partial w} , \frac{\partial L}{\partial b}$ to obtain the value of w and b
# hyperplane equation $\overrightarrow{w}\cdot\overrightarrow{x} +b$ is quite similar to line equation $mx +b$, which means b changes the intercept only

# $$\frac{\partial L}{\partial w} = \|\overrightarrow{w}\| - \displaystyle\sum^n_{i=1}\alpha_iy_ix_i = 0 \\
\|\overrightarrow{w}\| = \displaystyle\sum^n_{i=1}\alpha_iy_ix_i \\ 
\frac{\partial L}{\partial b} = -\displaystyle\sum^n_{i=1}\alpha_iy_i = 0$$

# Therefore, $$\\ L = \sum_i\alpha_i - \frac{1}{2}\sum_{i,j}\alpha_i\alpha_jy_iy_j\overrightarrow{x_i}\cdot\overrightarrow{x_j}$$ , where $\alpha^2$ is the quadratic problem, max L

# Optimization objective:
# 1. minimize $\|\overrightarrow{w}\|$ 2. maximize b for condition $y_i(\overrightarrow{x_i}\cdot\overrightarrow{w}+b)\geq 1$
# Class(knownfeatures$\cdot\overrightarrow{w}+b) \geq 1$  , w value behaves like an upward parabola and b behaves like an downward parabola (convex optimization). So they will have the optimized values at the tits
# Set initial value of w, find the respective b
# Continue to step down w, step up b to fulfill the condition, also need to test w = [a,b], [-a,b], [a,-b] and [-a,-b]. Also for x, [c,d], [c,-d],[-c,d],[-c,-d]
# Continue to find a lower value, until a greater value occurs, then smaller learning rate is needed to backtrack

# Minimum
# we want to find global but not local minimum, usually the problem is not complex so global is easily found.
# For the correct w and b, SV closes to 1
